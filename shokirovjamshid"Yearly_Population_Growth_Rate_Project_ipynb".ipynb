```python
def read_data(filename):
  """
  Read csv file
  Args:
    file_path(str): file path
  Returns:
    list: data split row
  """
  # WRITE YOUR CODE HERE
  f = open(filename,mode='r+')
  a = f.read()
  return a
print(read_data("Population.csv"))
```

```python
data_list = read_data("Population.csv")
```

```python
def get_data ( data_list ):
  a=data_list.split( '\n' )
  b = []
  c = []
  for i in  a:
    d =i.split(',')
    b.append(d[0])
    c.append(d[1])
    g = list(map(float,b[1:]))
    h = list(map(float,c[1:]))
  return g,h
get_data(data_list)
```

```python
labels, points = get_data(data_list)
```

```python
def plot_data( labels,points):
   plt.plot(labels,points,color= '#fc9403' )
   plt.xlabel( 'Yillar' ,color='green' )
   plt.ylabel("Aholi o'sishi" ,color= 'green' )
   plt.title( "Aholining yillik o'sish sur'ati" ,color= 'b' )
   plt.grid( 'majnor' )
   plt.show()
plot_data(labels,points)
```

```python
plt.figure(dpi=100)
plot_data(labels, points)
```
